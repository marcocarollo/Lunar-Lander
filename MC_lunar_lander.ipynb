#### GIF generation
Generate an episode with the trained agent and save it as a gif (there's *a bit* of initial selection bias to have a nicer gif).

```python
observation_space = (11, 7, 9, 9, 7, 7, 2, 2)
# test
env_test = gym.make('LunarLander-v2')
n_episodes = 100
performance_traj = np.zeros(n_episodes)
for i in range(n_episodes):
    
    state = env_test.reset(seed=i)[0]
    state = discretize_state(state, observation_space)
    done = False
    while not done:
        action = get_action_epsilon_greedy(MC.Qvalues, state, 0, 4)
        next_state, reward, truncated, terminated, _ = env_test.step(action)
        done = terminated or truncated
        performance_traj[i] += reward
        next_state = discretize_state(next_state, observation_space)
        state = next_state
best = np.argmax(performance_traj)
print("best episode: ", best)
print("best performance: ", performance_traj[best])
env_test.close()
```

```python
from PIL import Image

observation_space = (11, 7, 9, 9, 7, 7, 2, 2)
# test
env_test = gym.make('LunarLander-v2', render_mode="rgb_array")
frames = []
state = env_test.reset(seed=97)[0] #best episode found in the first 100 seeds
state = discretize_state(state, observation_space)
done = False
while not done:
    action = get_action_epsilon_greedy(MC.Qvalues, state, 0, 4) #change 0 to 1 to see random agent
    next_state, reward, truncated, terminated, _ = env_test.step(action)
    done = terminated or truncated
    next_state = discretize_state(next_state, observation_space)
    state = next_state
    frames.append(env_test.render())

env_test.close()
imgs = [Image.fromarray(img) for img in frames]
# duration is the number of milliseconds between frames; this is 40 frames per second
imgs[0].save("gifs/MC.gif", save_all=True, append_images=imgs[1:],duration = 50, loop=0)
```

#### $\varepsilon(t)$

```python
def epsilon(t):
    """
    This function returns the value of epsilon at count t.
    epsilon(t) = epsilon_0 / (1 + k * (t - t_star) ** 1.05)
    """
    return 0.2 / (1.0 + 0.00001 * (t-500000) ** 1.05)


#let's plot the function epsilon(t)
t = np.arange(0, 5000000, 100)
plt.plot(t, epsilon(t))
plt.xlabel("t")
plt.ylabel("epsilon")
plt.title("Epsilon function")
plt.show()
```
